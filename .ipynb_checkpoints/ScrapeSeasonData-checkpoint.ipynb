```python
import requests
from bs4 import BeautifulSoup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
from mplsoccer import Pitch, VerticalPitch
from statsbombpy import sb
import json
import ast
import seaborn as sns
jsonDec = json.decoder.JSONDecoder()
```

```python
base_url = 'https://understat.com/match/'
page = requests.get('https://understat.com/league/EPL/2019')
soup = BeautifulSoup(page.content, "html.parser")
scripts = soup.find_all('script')
strings = scripts[1].string
```

```python
# strip unnecessary symbols and get only JSON data 
ind_start = strings.index("('")+2 
ind_end = strings.index("')") 
json_data = strings[ind_start:ind_end] 
json_data = json_data.encode('utf8').decode('unicode_escape')

#convert string to json format
data = json.loads(json_data)
```

```python
url_list = []
for i in data:
    match = i['id']
    url = base_url+match
    url_list.append(url)
```

```python
len(url_list)
```

```python
df_complete_home = pd.DataFrame()
df_complete_away = pd.DataFrame()
for url in url_list:
    res = requests.get(url)
    soup = BeautifulSoup(res.content, 'lxml')
    scripts = soup.find_all('script')
    strings = scripts[1].string
    ind_start = strings.index("('")+2 
    ind_end = strings.index("')") 
    json_data = strings[ind_start:ind_end] 
    json_data = json_data.encode('utf8').decode('unicode_escape')

    #convert string to json format
    data = json.loads(json_data)
    df = pd.json_normalize(data['h'])
    df_complete_home = pd.concat([df_complete_home,df])
    df = pd.json_normalize(data['a'])
    df_complete_away = pd.concat([df_complete_away,df])
```

```python
fig ,ax = plt.subplots(figsize=(13.5,8))
fig.set_facecolor('#22312b')
ax.patch.set_facecolor('#22312b')
pitch = Pitch(pitch_type='statsbomb', orientation='horizontal',
              pitch_color='#22312b', line_color='#c7d5cc', figsize=(13, 8),
              constrained_layout=False, tight_layout=True)
pitch.draw(ax=ax)
plt.gca().invert_yaxis()
# kde = sns.kdeplot(
#         x=df['x'],
#         y=df['y'],
#         shade = True,
# #         shade_lowest=False,
#         alpha=.5,
#         n_levels=10,
#         cmap = 'magma'
# )
plt.xlim(0,120)
plt.ylim(0,80)
shot_result = df_complete_home['result'].values
for x in range(100):
#     print(shot_result[x])
    if shot_result[x] == 'Goal':
        plt.scatter(float(df_complete_home['X'].values[x])*120,float(df_complete_home['Y'].values[x])*80,color='green')
    else:
        plt.scatter(float(df_complete_home['X'].values[x])*120,float(df_complete_home['Y'].values[x])*80,color='red')
```
